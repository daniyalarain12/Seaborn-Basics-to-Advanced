# SEABORN + MATPLOTLIB TOGETHER

## Why Combine Both?
Seaborn and Matplotlib are designed to work together.
Seaborn provides beautiful high-level statistical plots with minimal code, while Matplotlib gives full control over figure layout, sizing, titles, labels and subplots.

## How They Work Together
- Use `plt.subplots()` to create the figure and axes grid — this is Matplotlib
- Pass `ax=ax[i]` to any Seaborn function to draw on a specific subplot
- Use `ax[i].set_title()`, `ax[i].set_xlabel()` etc. for per-subplot customization
- Use `fig.suptitle()` and `fig.tight_layout()` for figure-level control

## Key Rule
- **Seaborn** = creates the plot
- **Matplotlib** = controls the layout, size, titles and saving

```python
# Importing both Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Applying ticks theme for clean minimal styling
sns.set_theme(style='ticks')
```

```python
# Displaying all available built-in Seaborn datasets
sns.get_dataset_names()
```

```python
# Loading the tips dataset
df = sns.load_dataset('tips')
```

```python
# Creating a figure with 2 subplots side by side using Matplotlib
# figsize=(10,5) gives enough width for both plots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# LEFT SUBPLOT — Seaborn lineplot drawn on ax[0]
# Shows average bill amount for each day, split by gender
# errorbar=None removes confidence interval shading for a clean line
# marker='o' adds visible data points on the line
sns.lineplot(df, x='day', y='total_bill', hue='sex', ax=ax[0], marker='o', errorbar=None)
ax[0].set_title('AVERAGE BILL VALUE FOR EACH DAY')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Total Bill Amount')

# RIGHT SUBPLOT — Seaborn scatterplot drawn on ax[1]
# Shows relationship between total bill and tip, colored by gender
sns.scatterplot(df, x='total_bill', y='tip', hue='sex', ax=ax[1])
ax[1].set_title('RELATIONSHIP BETWEEN TOTAL BILL AND TIP')
ax[1].set_xlabel('Total Bill Amount')
ax[1].set_ylabel('Tip Amount')

# Adding a main title to the entire figure using Matplotlib
fig.suptitle('BILL TRENDS AND TIP ANALYSIS')

# Adjusting spacing between subplots to prevent overlapping
fig.tight_layout()
plt.show()
```
