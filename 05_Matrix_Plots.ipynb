# MATRIX PLOTS

- MATRIX PLOTS ARE USED TO VISUALIZE RECTANGULAR DATA IN A GRID FORMAT.
- THEY HELP IN UNDERSTANDING CORRELATION, PATTERNS, AND RELATIONSHIPS BETWEEN MULTIPLE VARIABLES.
- COMMONLY USED IN HEATMAPS, CORRELATION MATRICES, AND FEATURE RELATIONSHIP ANALYSIS.

sns.heatmap(data, cmap="color_map", annot=True, fmt="d", ax=ax) ---> USED TO CREATE A HEATMAP.

    - data ---> SPECIFIES THE 2D DATA STRUCTURE (SUCH AS A CORRELATION MATRIX OR PIVOT TABLE).
    - cmap="color_map" ---> DEFINES THE COLOR PALETTE USED TO REPRESENT VALUES (E.G., "viridis", "coolwarm", "magma").
    - annot=True ---> DISPLAYS NUMERICAL VALUES INSIDE EACH CELL OF THE HEATMAP.
    - fmt="d" ---> DEFINES THE STRING FORMATTING FOR ANNOTATED VALUES (E.G., INTEGER FORMAT).
    - ax=ax ---> DRAWS THE PLOT ON A SPECIFIC AXES OBJECT FOR USE IN SUBPLOTS.

```python
# Importing Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Applying ticks style theme
sns.set_theme(style='ticks')
```

```python
# Displaying all available built-in Seaborn datasets
sns.get_dataset_names()
```

```python
# Loading the flights dataset
# Contains monthly airline passenger counts from 1949 to 1960
df = sns.load_dataset('flights')
```

```python
# Previewing the raw flights dataset before pivoting
df.head()
```

```python
# Pivoting the dataset to create a 2D matrix:
# Rows = months, Columns = years, Values = passenger counts
# This format is required for heatmap visualization
df = df.pivot(index='month', columns='year', values='passengers')
```

```python
# Basic heatmap using coolwarm colormap
# Darker red = higher passengers, darker blue = lower passengers
sns.heatmap(df, cmap='coolwarm')
plt.show()
```

```python
# annot=True displays the actual passenger count inside each cell
# fmt='d' formats the values as integers (no decimal places)
sns.heatmap(df, cmap='coolwarm', annot=True, fmt='d')
plt.title('Passengers Heatmap')
plt.tight_layout()
plt.show()
```
