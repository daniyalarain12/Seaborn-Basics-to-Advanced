# DISTRIBUTION PLOTS

Distribution plots are used to visualize the distribution of a single numerical variable.
They show how data is spread across different value ranges.

sns.histplot(data=dataframe, x="column_name", bins=value, ax=ax) ---> USED TO CREATE A HISTOGRAM.
- data=dataframe ---> SPECIFIES THE DATAFRAME CONTAINING THE DATA.
- x="column_name" ---> SPECIFIES THE COLUMN WHOSE DISTRIBUTION IS TO BE PLOTTED.
- bins=value ---> DEFINES THE NUMBER OF BINS TO DIVIDE THE DATA INTO.
- ax=ax ---> DRAWS THE PLOT ON A SPECIFIC AXES OBJECT FOR USE IN SUBPLOTS.

```python
# Importing Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Applying ticks style theme
sns.set_theme(style='ticks')
```

```python
# Displaying all available built-in Seaborn datasets
sns.get_dataset_names()
```

```python
# Loading the penguins dataset
# Contains physical measurements of 3 penguin species
df = sns.load_dataset('penguins')
```

```python
# Histogram showing distribution of penguin body mass
# bins=20 divides the data into 20 equal intervals
# Shows how body mass is distributed across all penguins in the dataset
sns.histplot(df, x='body_mass_g', bins=20)
plt.show()
```
