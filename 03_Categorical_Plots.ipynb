# CATEGORICAL PLOTS

Categorical plots are used to visualize the distribution or relationship of a numerical variable across different categories.

- sns.barplot(data=dataframe, x="column_name", y="column_name", hue="column_name", ax=ax) ---> USED TO CREATE A BAR CHART.
- sns.boxplot(data=dataframe, x="column_name", y="column_name", hue="column_name", ax=ax) ---> USED TO CREATE A BOX CHART.

```python
# Importing Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Applying ticks style theme
sns.set_theme(style='ticks')
```

```python
# Displaying all available built-in Seaborn datasets
sns.get_dataset_names()
```

```python
# Loading the tips dataset
df = sns.load_dataset('tips')
```

```python
# Bar plot showing average tip amount for each day of the week
# Seaborn automatically calculates the mean and adds confidence intervals
sns.barplot(data=df, x='day', y='tip')
plt.show()
```

```python
# Adding hue='sex' to compare tipping behavior between Male and Female
# Creates grouped bars — one for each gender per day
sns.barplot(data=df, x='day', y='tip', hue='sex')
plt.show()
```

```python
# Box plot showing tip distribution for each day
# Shows median, Q1, Q3, whiskers and outliers for each day
sns.boxplot(data=df, x='day', y='tip')
plt.show()
```

```python
# Adding hue='sex' to compare tip distribution between Male and Female per day
sns.boxplot(data=df, x='day', y='tip', hue='sex')
plt.show()
```
