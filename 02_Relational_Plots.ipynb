# RELATIONAL PLOTS

Axes-level relational plots — work on a single axes object and can be combined with Matplotlib subplots.
These are the axes-level equivalents of relplot().

- sns.scatterplot(data=dataframe, x="column_name", y="column_name", hue="column_name", size="column_name", style="column_name", ax=ax) ---> USED TO CREATE A SCATTER PLOT.
- sns.lineplot(data=dataframe, x="column_name", y="column_name", errorbar=None, ax=ax) ---> USED TO CREATE A LINE PLOT.
    - errorbar=None ---> DISABLES THE DISPLAY OF ERROR BARS OR CONFIDENCE INTERVALS AROUND THE LINE.
    - ax=ax ---> DRAWS THE PLOT ON A SPECIFIC AXES OBJECT. COMMONLY USED WHEN WORKING WITH SUBPLOTS.

```python
# Importing Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Applying ticks style theme for clean minimal look
sns.set_theme(style='ticks')
```

```python
# Displaying all available built-in Seaborn datasets
sns.get_dataset_names()
```

```python
# Loading tips dataset and creating a scatter plot
# hue='time' colors points by Lunch or Dinner — adds a third dimension
tips = sns.load_dataset('tips')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.show()
```

```python
# Previewing the tips dataset
tips.head()
```

```python
# Loading flights dataset — contains monthly passenger counts from 1949 to 1960
# Basic line plot showing passenger growth trend over years
flights = sns.load_dataset('flights')
sns.lineplot(data=flights, x='year', y='passengers')
plt.show()
```

```python
# errorbar=None removes the confidence interval shading around the line
# Gives a cleaner line without the shaded band
flights = sns.load_dataset('flights')
sns.lineplot(data=flights, x='year', y='passengers', errorbar=None)
plt.show()
```
